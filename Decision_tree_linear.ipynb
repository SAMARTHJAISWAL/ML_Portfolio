```python
%matplotlib inline
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
patients = pd.read_csv('linear_regression_dataset.csv')
patients = patients.dropna()
```

```python
pt =patients.iloc[:349,:]
test = patients.iloc[349:,:]
```

```python
X = patients[['AGE','FEMALE', 'LOS', 'APRDRG', ]]

Y = patients[['TOTCHG']]
```

```python
X_test = test[[ 'AGE', 'FEMALE', 'LOS', 'APRDRG',]]

Y_test = test[['TOTCHG']]
```

```python
X = X.values
Y = Y.values
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
scalar.fit(X)
scalar2 = StandardScaler()
scalar2.fit(X_test)




scaled_data = scalar.transform(X)
scaled_tdata = scalar2.transform(X_test)

#from sklearn.decomposition import PCA
#pca = PCA(n_components = 2)
#pca2 = PCA(n_components = 2)
#pca.fit(scaled_data)
#pca2.fit(scaled_tdata)
#x_pca = pca.transform(scaled_data)
#t_pca = pca2.transform(scaled_tdata)
```

```python
 from sklearn.tree import DecisionTreeRegressor
```

```python
DtReg = DecisionTreeRegressor(ccp_alpha=0.001)
```

```python
DtReg.fit(X, Y)
```

```python
y_pred_t = DtReg.predict((X_test))
```

```python
from sklearn import metrics
```

```python
r_square = metrics.r2_score(Y_test, y_pred_t)
print('R-Square Error associated with Decision Tree Regression is:', r_square)
```
